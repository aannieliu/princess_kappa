In this notebook, we are trying to do some exploration on all the movies we predicted to win. By all we mean all the movies from our logistics and random forest model and there are total of 34 movies. 

We first create a list called win_list that contains all the predicted movie names

```python
win_list = ['Jurassic World', 'Mad Max: Fury Road', 'Straight Outta Compton',
'The Revenant',
'Spotlight',
'Spectre',
'Mission: Impossible - Rogue Nation',
'Avengers: Age of Ultron',
'Furious 7',
'The Hateful Eight',
'The Martian',
'Bridge of Spies',
'Muhammad: The Messenger of God',
'Dilwale',
'Ex Machina',
'Child 44',
'Black Mass',
'Straight Outta Compton',
'Fantastic Four',
'Vacation',
'Anomalisa',
'Creed',
'Sleeping with Other People',
'Room',
'The Diary of a Teenage Girl',
'Insidious: Chapter 3',
'Captive',
'American Ultra',
'The Walk',
'Son of Saul',
'Amnesia',
'Muhammad: The Messenger of God',
'The Gift',
'Embrace of the Serpent',
'Jackpot',
'7 Letters']
```

```python
ultimatedf = pd.read_csv('final_table2015.csv')
ultimatedf.shape
```

Then we subset the predicted movies from the original table.

```python
win_df = ultimatedf[ultimatedf.Title.isin(win_list)]
```

```python
win_df.shape
```

We start our analysis by making some plots.

**Fig.1** shows the the histogram of the runtime of all movies. We can see that the predicted movies have a wide range of runtime, and there are 11 of them have about 100 miniutes of playtime. The mean runtime is 125 miniutes.

```python
plt.hist(win_df.Runtime, color='palevioletred',bins=np.arange(50,250,5));
plt.title('Fig 1. Histogram of predicted movie runtime');
plt.axvline(np.mean(win_df.Runtime), color='b', linewidth=3, label='mean of all')
#plt.axvline(np.median(win_df.Runtime), color='g', linewidth=3, label='median of all');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
print np.mean(win_df.Runtime)
```

Next, we plot the histogram of box office. **Fig 2** shows the box office of all movies.  **Fig 3** shows the box office of all movies with box office less than 1 billion. We see that most movies have box office less than 200 million USD. 

```python
plt.figure(1)
plt.subplot(211)
dt = win_df.box_USD.dropna().values/1000000
binwidth = 180
plt.hist(dt, bins=np.arange(0, 700, 50), color='salmon');
#plt.axvline(np.mean(ultimatedf.box_USD.dropna())/1000000, color='b', linewidth=3, label='mean')
#plt.axvline(np.median(ultimatedf.box_USD.dropna())/1000000, color='g', linewidth=3, label='median')
plt.title('Fig.2 Histogram of box office in USD (million)');

plt.subplot(212)
dt = win_df.box_USD.dropna().values/1000000
binwidth = 15
dt_1=dt[dt<=1000]
plt.hist(dt_1, bins=np.arange(min(dt_1), max(dt_1) + binwidth, binwidth), color='tomato');
#plt.axvline(np.mean(ultimatedf.box_USD.dropna())/1000000, color='b', linewidth=3, label='mean')
#plt.axvline(np.median(ultimatedf.box_USD.dropna())/1000000, color='g', linewidth=3, label='median')
plt.title('Fig.3 Box office of predicted movies with less than 1 billion');
plt.xlabel('box office in USD(millions)');
```

**Fig 4** plots four histograms. The first one is mean of the actors' age at the time of their nominations. The second one is the directors' age at the time of their nominations. The third shows the number of movies the actors are in upto now. The fourth shows the number of movies the directors directed upto now. All plots show approximately normal distribution except mean director credit where most of the value concentrated under 10.

```python
plt.figure(figsize=(15,15))
ax1 = plt.subplot(221)
ax1.hist(win_df.A_mean_age)
ax1.set_title('Fig.4 Histogram of mean actor age')
ax1.set_xlabel('mean age')
ax1.set_ylabel('count')

ax2 = plt.subplot(222)
ax2.hist(win_df.D_mean_age)
ax2.set_title('Histogram of mean director age')
ax2.set_xlabel('mean age')
ax2.set_ylabel('count');

ax3 = plt.subplot(223)
ax3.hist(win_df.A_mean_credit)
ax3.set_title('Histogram of mean actor credit')
ax3.set_xlabel('mean credit')
ax3.set_ylabel('count')

ax4 = plt.subplot(224)
ax4.hist(win_df.D_mean_credit)
ax4.set_title('Histogram of mean director credit')
ax4.set_xlabel('mean credit')
ax4.set_ylabel('count');
```

**Fig 5** shows the percentage of genres for all the predicted movies. The percentage of each genre is very similar to that calculated from all the movies (both winning and nominated). The genres with highest percentage is still 'Drama'. However, 'Romance', which was ranked second previously, has a low ranking now.

```python
genrefrac = win_df.mean(axis=0)*100
## Only get the genres
genrefrac = genrefrac[17:37]
genrefrac.sort()
genrefrac.plot(kind='barh', color='hotpink').set_ylabel("Genres");
plt.xlabel('Percentage');
plt.title('Fig.5 Percentage of genres for predicted movies')
plt.legend().set_visible(False);
```

At last, we look at the popularity of all the predicted movies. Popularity of these movies can be indicated by their number of reviews on rotten tomatos and IMDB. Based on the two histograms below, we can see that most of the predicted movies have number of reviews under 100000, and there are only a few with more than 200000 reviews.

```python
plt.hist(win_df.tomatoUserReviews[~np.isnan(win_df.tomatoUserReviews)], color='palevioletred');
plt.title('Fig 6. Histogram of predicted movie tomatoReviews');
plt.axvline(np.mean(win_df.tomatoUserReviews), color='b', linewidth=3, label='mean')
#plt.axvline(np.median(win_df.Runtime), color='g', linewidth=3, label='median of all');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
print np.mean(win_df.tomatoUserReviews)
```

```python
plt.hist(win_df.imdbVotes[~np.isnan(win_df.imdbVotes)], color='palevioletred');
plt.title('Fig 7. Histogram of predicted movie imdbVotes');
plt.axvline(np.mean(win_df.imdbVotes), color='b', linewidth=3, label='mean')
#plt.axvline(np.median(win_df.Runtime), color='g', linewidth=3, label='median of all');
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
print np.mean(win_df.imdbVotes)
```
